# Теоретические основы

## Информационная энтропия

Возьмем два случайных события $x$, распределенные с вероятностями $\{p_1, p_2, ..., p_n\} между $n$ состояниями.

Двоичная энтропия этих событий записывается как 

$$H(x) = -\sum\limits_{1}^{n} p_i log_2(p_i)$$

Но что это значит?

Прирост ифнормации связан с уменьшением неопределенности. Клод Шеннон предложил измерять количесвто информации:
- мера должна быть непрерывной (функцией)
- если все варианты равновероятны, увеличение количества вариантов должно приводить к увеличению энтропии
- если "разделить" распределение, то сумма функций энтропии распределений должна равняться энтропии полного распределения

Всем этим требованиям удовлетворяет единственная функция: 

$$-K\sum\limits_{1}^{n} p_i log_2(p_i)$$

$$H(X) = E[I(X)] = E[-log(P(X))]$$
